```python
#Importar librerias
import pandas as pd
import numpy as np

# Cargar el archivo Excel
archivo_excel = '/content/EHR.xlsx'
#Seleccionar las hojas de calculo que contiene los datos clinicos y los procedimientos realizados
informacion_base = pd.read_excel(archivo_excel, sheet_name='InformacionBase')
procedimientos = pd.read_excel(archivo_excel, sheet_name='ProcedimientosOrdenadosdeBase')
# Agregar la columna ID en InformacionBase incluyendo el folio de atencion
informacion_base['ID'] = informacion_base['HistoriaClinica'].astype(str) + informacion_base['Folio'].astype(str)
# Agregar la columna ID en ProcedimientosOrdenadosdeBase
procedimientos['ID'] = procedimientos['HistoriaClinica'].astype(str) + procedimientos['Folio'].astype(str)
# Concatenar los procedimientos por ID y eliminar la columnas innecesarias
procedimientos['procedimientos'] = procedimientos.groupby('ID')['DescripcionProcedumiento'].transform(lambda x: ', '.join(x) + '.')
procedimientos = procedimientos.drop_duplicates(subset='ID')
procedimientos = procedimientos.drop(columns=['HistoriaClinica', 'Folio', 'DescripcionProcedumiento'])
informacion_base = informacion_base.drop(columns=['HistoriaClinica', 'Folio', 'MotivoConsulta'])
df_final = informacion_base.merge(procedimientos, on='ID', how='left')
# En caso de no tener procedimientos registrados se incluye ese dato de forma explicita
df_final['procedimientos'].fillna("NO REGISTRA FORMULACION ADICIONAL", inplace=True)
# Crear la columna DATOS concatenando las columnas requeridas en minúsculas
df_final['DATOS'] = (
    "EDAD: " + df_final['EdadAtencion'].astype(str).str.lower() + ".\n\n" +
    "ENFERMEDAD ACTUAL: " + df_final['EnfermedadActual'].astype(str).str.lower() + ".\n\n" +
    "PROCEDIMIENTOS / FORMULACION EN SISTEMA: " + df_final['procedimientos'].astype(str).str.lower() + ".\n\n"
)
# Conservar solo las columnas 'ID' y 'DATOS'
df_final = df_final[['ID', 'DATOS']]
# Generar números aleatorios para anonimizar los IDs
df_final['ID'] = (np.random.randint(100000, 999999, size=(len(df_final))))
df_final.to_excel('/content/EHRFINAL.xlsx', index=False)
```
